6-Explain the concept of generators in Python and how they are defined.

   ->Generators in Python are a type of iterable, like lists or tuples, that allow you to iterate over data one item at a time, lazily. Instead of computing all the values at once and storing them in memory, generators produce items on the fly. This makes them more memory-efficient when working with large datasets or streams of data.

How Generators Work

Generators are defined using functions but utilize the yield statement instead of return. When a generator function is called, it doesn't execute its body immediately. Instead, it returns a generator object, which can be iterated over. Each time the yield statement is executed, the function's state is "paused," allowing it to resume from that point on the next iteration.

10-  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  ->The map(), reduce(), and filter() functions in Python are higher-order functions commonly used for functional programming tasks. They each serve a distinct purpose in processing data collections.

1. map()

The map() function applies a given function to each item in an iterable (e.g., list, tuple) and returns a new iterable with the results.

Syntax:

map(function, iterable)

Function: A callable that operates on individual elements of the iterable.

Iterable: The data collection to process.


Key Features:

Transforms each item in the iterable.

Returns a map object, which can be converted to a list, tuple, etc.


Example:

# Doubling each number in a list
nums = [1, 2, 3, 4]
result = map(lambda x: x * 2, nums)
print(list(result))  # Output: [2, 4, 6, 8]


2. reduce()

The reduce() function, from the functools module, applies a binary function cumulatively to the elements of an iterable, reducing it to a single value.

Syntax:

reduce(function, iterable[, initializer])

Function: A callable that takes two arguments and combines them.

Iterable: The data collection to process.

Initializer (optional): A starting value for the reduction.


Key Features:

Performs cumulative computations like summation, product, etc.

Returns a single result.


Example:

from functools import reduce

# Computing the product of a list
nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x * y, nums)
print(result)  # Output: 24

3. filter()

The filter() function applies a predicate (a function returning True or False) to each element of an iterable and includes only the elements that satisfy the condition.

Syntax:

filter(function, iterable)

Function: A callable that returns True or False.

Iterable: The data collection to filter.


Key Features:

Filters elements based on a condition.

Returns a filter object, which can be converted to a list, tuple, etc.


Example:

# Filtering out even numbers
nums = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, nums)
print(list(result))  # Output: [2, 4]

Key Differences

By understanding these distinctions, you can choose the right tool for tasks involving data transformation, aggregation, or filtering.

1- Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.


```python
def sum_of_evens(numbers):
    """
    Returns the sum of all even numbers in the given list.

    :param numbers: List of numbers
    :return: Sum of even numbers
    """
    return sum(num for num in
               numbers if num % 2 == 0)

# Example usage:
numbers_list = [1, 2, 3, 4, 5, 6]
result = sum_of_evens(numbers_list)
print(f"sum_of_evens numbers {result}") # Output: 12

```

2-Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(input_string):
    """
    Reverses the given string and returns it.

    Args:
        input_string (str): The string to be reversed.

    Returns:
        str: The reversed string.
    """
    return input_string[::-1]

# Example usage
reversed_string = reverse_string("Hello, World!")
print(reversed_string)  # This function uses Python's slicing technique ([::-1]) to reverse the string efficiently.
```

4-Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    """Check if a given number is prime."""
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Test the function for numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")# we can add this also else:print(f"{num} is not a prime number")
```

5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

  ->Here is an implementation of an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms:

```python
class FibonacciIterator:
    def __init__(self, n): # changed _init_ to __init__
        """
        Initialize the iterator with the specified number of terms.
        :param n: The number of terms to generate.
        """
        self.n = n  # Total number of terms
        self.current = 0  # Current position in the sequence
        self.a, self.b = 0, 1  # Initial values for Fibonacci sequence

    def __iter__(self): # changed _iter_ to __iter__
        """
        Returns the iterator object itself.
        """
        return self

    def __next__(self): # changed _next_ to __next__
        """
        Returns the next number in the Fibonacci sequence.
        """
        if self.current >= self.n:
            raise StopIteration  # Stop the iteration if we've reached the limit

        # Compute the next Fibonacci number
        if self.current == 0:
            self.current += 1
            return self.a
        elif self.current == 1:
            self.current += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.current += 1
            return self.a

# Example usage
n = 10
fib = FibonacciIterator(n) # Create an instance of the FibonacciIterator
for num in fib:
    print(num)
```

6- Write a generator function in Python that yields the powers of 2 up to a given exponent.

  ->Here’s a generator function in Python that yields the powers of 2 up to a given exponent:

```python
def powers_of_two(max_exponent):
    """
    A generator function that yields powers of 2 up to 2**max_exponent.

    :param max_exponent: The maximum exponent (inclusive) for 2**n.
    """
    for n in range(max_exponent + 1):
        yield 2 ** n

# Example usage:
for power in powers_of_two(5):
    print(power)
```

7- Implement a generator function that reads a file line by line and yields each line as a string

  -> Here’s a Python implementation of a generator function that reads a file line by line and yields each line as a string:

```python
def read_file_line_by_line(file_path):
    """
    Generator function to read a file line by line.

    :param file_path: Path to the file to be read
    :yield: Each line of the file as a string
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Stripping newline characters
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
    except IOError as e:
        print(f"Error reading file: {e}")

file_path = 'example.txt'

for line in read_file_line_by_line(file_path):
    print(line)
```

10- Create a Python program that uses `filter()` to remove all the vowels from a given string.

  ->Here is a Python program that uses the filter() function to remove all vowels from a given string

```python
def remove_vowels(input_string):
    # Define a function that returns True if the character is not a vowel
    def is_not_vowel(char):
        return char.lower() not in 'aeiou'

    # Use filter() to remove vowels from the input string
    filtered_chars = filter(is_not_vowel, input_string)

    # Join the filtered characters back into a string
    result = ''.join(filtered_chars)
    return result

# Example usage
input_string = "Hello, World!"
output_string = remove_vowels(input_string)
print(f"Original: {input_string}")
print(f"Without vowels: {output_string}")
```
